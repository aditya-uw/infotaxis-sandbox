```python
n = 100
test = np.ones(n)
test[int(n/2):] = 1 + 1e-16
unique, counts = np.unique(test, return_counts=True)
assert((counts == n).all())
unique, counts
```

```python
aux = np.array([1, 1, 1 + 1e-15, 1 + 1e-14])
aux.sort()
aux
```

```python
mask = np.empty(aux.shape, dtype=np.bool)
mask[:1] = True
mask
```

```python
aux_firstnan = np.searchsorted(aux, aux[-1], side='left')
aux_firstnan
```

```python
mask[aux_firstnan] = True
mask[aux_firstnan + 1:] = False
mask
```

```python
mask[1:] = aux[1:] != aux[:-1]
mask
```

```python
ret = (aux[mask],)
ret
```

```python
np.nonzero(mask) + ([mask.size],)
```
